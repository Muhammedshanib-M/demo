```python
y=df['SPENDING SCORE(1-100)']
x=df.drop(columns=['SPENDING SCORE(1-100)'])
y
x
```

importing train_test_split library

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,)

#print the shaps of the resulting datasets to verity the split
print("x_train shape:",x_train.shape)
print("x_test shape:",x_test.shape)
print("y_train shape:",y_train.shape)
print("y_test shape:",y_test.shape)
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
labelencoder = LabelEncoder()

# Encode the 'GENDER' column in x_train
x_train['GENDER'] = labelencoder.fit_transform(x_train['GENDER'])

# Encode the 'GENDER' column in x_test
x_test['GENDER'] = labelencoder.transform(x_test['GENDER'])

# Print the encoded data
print(x_train)
print(x_test)
```

```python
# prompt: support vector machine regressor

from sklearn.svm import SVR

# Create a Support Vector Regressor object
svr = SVR(kernel='sigmoid')

# Fit the model to the training data
svr.fit(x_train, y_train)

# Predict on the test data
y_pred = svr.predict(x_test)

# Print the predicted values
print(y_pred)


```

```python
from sklearn.metrics import mean_absolute_error

# Calculate the mean absolute error
svm_mae = mean_absolute_error(y_test, y_pred)

# Print the result
print("Mean absolute error:", svm_mae)
```

```python

from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
from math import sqrt

# Assuming you have your training data X_train, y_train and test data X_test, y_test

# Train your SVM model
svm_model = SVR(kernel='linear')
svm_model.fit(x_train, y_train)

# Make predictions on the test data
svm_prediction = svm_model.predict(x_test)

# Calculate the mean squared error
mean_sqrd_error = mean_squared_error(y_test, svm_prediction)

# Calculate the root mean squared error
RMSE = sqrt(mean_sqrd_error)

# Print or use the MSE and RMSE
print("Mean Squared Error:", mean_sqrd_error)
print("Root Mean Squared Error:", RMSE)
```
